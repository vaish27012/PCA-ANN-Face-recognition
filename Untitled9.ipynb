```python
# PCA + ANN Face Recognition using LFW Dataset


import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Load LFW dataset
faces = fetch_lfw_people(
    min_faces_per_person=70,
    resize=1.0,
    color=False
)

X = faces.data
y = faces.target
names = faces.target_names
images = faces.images

h, w = images.shape[1], images.shape[2]

print("Total Images:", X.shape[0])
print("Features:", X.shape[1])
print("Classes:", names)

# 2. Show one input image from each class
n_classes = len(names)

fig, axes = plt.subplots(2, 4, figsize=(14, 8))
axes = axes.ravel()

for class_id in range(n_classes):
    index = np.where(y == class_id)[0][0]

    axes[class_id].imshow(images[index], cmap="gray", interpolation="bicubic")
    axes[class_id].set_title(f"Input: {names[class_id]}", fontsize=10)
    axes[class_id].axis("off")

for j in range(n_classes, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()

# 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

# 4. PCA feature extraction
n_components = 180

pca = PCA(
    n_components=n_components,
    whiten=True,
    random_state=42
)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# 5. Show eigenfaces
eigenfaces = pca.components_.reshape((n_components, h, w))

fig, axes = plt.subplots(3, 4, figsize=(10, 8))
axes = axes.ravel()

for i in range(12):
    axes[i].imshow(eigenfaces[i], cmap="gray")
    axes[i].set_title(f"eigenface {i}", fontsize=11)
    axes[i].axis("off")

plt.tight_layout()
plt.show()

# 6. Train ANN / MLP classifier
model = MLPClassifier(
    hidden_layer_sizes=(512, 256),
    activation="relu",
    solver="adam",
    alpha=0.0001,
    learning_rate="adaptive",
    max_iter=1500,
    random_state=42
)

model.fit(X_train_pca, y_train)

# 7. Prediction
y_pred = model.predict(X_test_pca)
y_prob = model.predict_proba(X_test_pca)

accuracy = accuracy_score(y_test, y_pred)

print("\nAccuracy:", round(accuracy * 100, 2), "%")

print("\nClassification Report:")
print(classification_report(
    y_test,
    y_pred,
    target_names=names,
    zero_division=1
))

# 8. Show one output image from each true class
fig, axes = plt.subplots(2, 4, figsize=(14, 9))
axes = axes.ravel()

shown_classes = set()
plot_index = 0

for i in range(len(X_test)):
    true_class = y_test[i]

    if true_class not in shown_classes:
        axes[plot_index].imshow(
            X_test[i].reshape(h, w),
            cmap="gray",
            interpolation="bicubic"
        )

        pred_name = names[y_pred[i]]
        true_name = names[y_test[i]]
        confidence = np.max(y_prob[i]) * 100

        axes[plot_index].set_title(
            f"Pred: {pred_name}\n"
            f"Prob: {confidence:.1f}%\n"
            f"True: {true_name}",
            fontsize=10
        )

        axes[plot_index].axis("off")

        shown_classes.add(true_class)
        plot_index += 1

    if plot_index == n_classes:
        break

for j in range(plot_index, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()
```
